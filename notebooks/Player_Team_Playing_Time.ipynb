```python
import pandas as pd
import numpy as np
```

```python
pl_time = pd.read_fwf('player_minutes_2425.txt')
pl_time = pl_time.iloc[:, 0]
pl_time[len(pl_time)] = '='

row_sep = [i for i in range(len(pl_time)) if '=' in pl_time.iloc[i]]
title_team_index = [0] + [i+1 for i in row_sep[:-1] if (' (' not in pl_time.iloc[i+1]) & ('=' not in pl_time.iloc[i+1]) & (':' not in pl_time.iloc[i+1])]
#title_team_index += [len(pl_time) - 1]
for i_title_team_index in range(len(title_team_index)):
    #print (i_title_team_index, pl_time.iloc[title_team_index[i_title_team_index]])
    if i_title_team_index < 19:
        title_team_list = list(pl_time.iloc[title_team_index[i_title_team_index] + 1:title_team_index[i_title_team_index + 1]])
        # title_row_sep = [i + 1 for i in row_sep if (title_team_index[i_title_team_index] < i) & (i < title_team_index[i_title_team_index + 1])]
    elif i_title_team_index == 19:
        title_team_list = list(pl_time.iloc[title_team_index[i_title_team_index] + 1:])
    #     title_row_sep = [i + 1 for i in row_sep if (title_team_index[i_title_team_index] < i)]
    # else:
    #     print ('Problem')

    title_row_sep = [i for i in range(len(title_team_list)) if '=' in title_team_list[i]]
    
    all_players = np.unique([name.split(':')[0] for name in title_team_list if ':' in name])
    df_player_time = pd.DataFrame({})
    df_player_time['vs'] = []
    df_player_time['loc'] = []
    for name in all_players:
        df_player_time[name] = []

    #print (len(title_row_sep))
    for i_title_row_set in range(len(title_row_sep) - 1):
        #print (i_title_row_set)

        
        game_list = list(title_team_list[title_row_sep[i_title_row_set] + 2:title_row_sep[i_title_row_set + 1]])
        #print (len(game_list))
        #print (title_team_list[title_row_sep[i_title_row_set]+1].split(' ')[0])
        
        remove_player = []
        for i_game_player in range(len(game_list)):
            if ':' not in game_list[i_game_player]:
                remove_player.append(i_game_player)

        for i_remove in np.sort(remove_player)[::-1]:
            del game_list[i_remove]
        
        game_players = [gl.split(':')[0] for gl in game_list]
        game_min = [float(g_min) if (' ' in g_min) else 0 for g_min in [gl.split(':')[1] for gl in game_list]]
        
        player_min_game = []
        player_min_game.append(title_team_list[title_row_sep[i_title_row_set]+1].split(' ')[0])
        player_min_game.append(title_team_list[title_row_sep[i_title_row_set]+1].split(' (')[1][:-1])
        
        for player in all_players:
            if player in game_players:
                player_min_game.append(game_min[np.argwhere(np.array(game_players) == player)[0][0]])
            else:
                player_min_game.append(np.nan)
        df_player_time.loc[len(df_player_time)] = player_min_game
    df_player_time.to_csv('Player_Time/' + pl_time.iloc[title_team_index[i_title_team_index]] + '.csv')
```
